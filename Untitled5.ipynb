## Deep Learning Model for Image Denoising using Autoencoder on MNIST

This notebook will guide you through building a convolutional autoencoder to remove noise from MNIST handwritten digit images.

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
import numpy as np
import matplotlib.pyplot as plt
```

### Load and Preprocess the MNIST Dataset

We will load the MNIST dataset, normalize the pixel values, and reshape the images to include the channel dimension.

```python
# Load MNIST dataset
(x_train, _), (x_test, _) = tf.keras.datasets.mnist.load_data()

# Normalize and reshape the data
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Reshape to (number, height, width, channel)
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))
x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))

print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
```

### Add Noise to Images

We will add random Gaussian noise to the `x_train` and `x_test` datasets to create `x_train_noisy` and `x_test_noisy`. These noisy images will be the input to our autoencoder, and the original images will be the target output.

```python
# Define noise factor
noise_factor = 0.5

# Add noise to training data
x_train_noisy = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape)
x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)

# Clip pixel values to stay within [0, 1] range
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)

print(f"x_train_noisy shape: {x_train_noisy.shape}")
print(f"x_test_noisy shape: {x_test_noisy.shape}")
```

### Build the Autoencoder Model

We will construct a convolutional autoencoder. The encoder will progressively reduce the spatial dimensions and increase feature maps, while the decoder will upsample and reduce feature maps to reconstruct the image.

```python
# Input shape (28, 28, 1)
input_img = Input(shape=(28, 28, 1))

# Encoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = tf.keras.layers.BatchNormalization()(x)
x = MaxPooling2D((2, 2), padding='same')(x) # (14, 14, 32)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = tf.keras.layers.BatchNormalization()(x)
encoded = MaxPooling2D((2, 2), padding='same')(x) # (7, 7, 32)

# Decoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(encoded)
x = tf.keras.layers.BatchNormalization()(x)
x = UpSampling2D((2, 2))(x) # (14, 14, 32)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = tf.keras.layers.BatchNormalization()(x)
x = UpSampling2D((2, 2))(x) # (28, 28, 32)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x) # (28, 28, 1)

# Autoencoder model
autoencoder = Model(input_img, decoded)

# Compile the autoencoder
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

autoencoder.summary()
```

### Train the Autoencoder

We will train the autoencoder model using the noisy training images (`x_train_noisy`) as input and the original clean training images (`x_train`) as the target output.

```python
# Train the autoencoder
history = autoencoder.fit(x_train_noisy, x_train,
                    epochs=15, # Changed epochs to 15
                    batch_size=128,
                    shuffle=True,
                    validation_data=(x_test_noisy, x_test))

# Plot training & validation loss values
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.show()
```

### Visualize Denoising Results

Let's visualize how well our autoencoder performs at denoising. We will take a few images from the test set, add noise to them, and then use our trained autoencoder to reconstruct the denoised versions. Finally, we'll display the original, noisy, and denoised images side-by-side for comparison.

```python
# Predict on test images
decoded_imgs = autoencoder.predict(x_test_noisy)

# Number of images to display
n = 10
plt.figure(figsize=(20, 4))
for i in range(n):
    # Display original
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    if i == 0:
        ax.set_title('Original')

    # Display noisy
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    if i == 0:
        ax.set_title('Noisy')

    # Display reconstruction
    ax = plt.subplot(3, n, i + 1 + 2 * n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    if i == 0:
        ax.set_title('Denoised')
plt.show()
```

### Train the Autoencoder

We will train the autoencoder model using the noisy training images (`x_train_noisy`) as input and the original clean training images (`x_train`) as the target output.

```python
# Train the autoencoder
history = autoencoder.fit(x_train_noisy, x_train,
                    epochs=15,
                    batch_size=128,
                    shuffle=True,
                    validation_data=(x_test_noisy, x_test))

# Plot training & validation loss values
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.show()
```
